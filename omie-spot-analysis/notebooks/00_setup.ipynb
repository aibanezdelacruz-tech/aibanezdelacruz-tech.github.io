## 1. Validación de dependencias

```python
import subprocess
import sys
from pathlib import Path

# Dependencias requeridas
required_packages = {
    'pandas': '>=1.5.0',
    'numpy': '>=1.23.0',
    'xgboost': '>=1.7.0',
    'scikit-learn': '>=1.1.0',
}

print("Validando dependencias...")
print("="*50)

missing = []
for pkg in required_packages.keys():
    try:
        __import__(pkg)
        print(f"[OK] {pkg}")
    except ImportError:
        print(f"[FAIL] {pkg} NO INSTALADO")
        missing.append(pkg)

if missing:
    print(f"\n[ERROR] Instala los paquetes faltantes:")
    print(f"  pip install {' '.join(missing)}")
else:
    print("\n[OK] Todas las dependencias listos")
```

## 2. Validación de estructura de carpetas

```python
project_root = Path('..')

required_dirs = [
    'data/raw/2023-2025',
    'data/raw/esios',
    'data/processed',
    'src',
    'notebooks',
    'reports/figures',
    'docs',
]

print("Validando estructura de carpetas...")
print("="*50)

for dir_path in required_dirs:
    full_path = project_root / dir_path
    status = "[OK]" if full_path.exists() else "[FAIL]"
    print(f"{status} {dir_path}")

print("\n[OK] Estructura de carpetas correcta")
```

## 3. Validación de datos OMIE

```python
from datetime import datetime

raw_dir = project_root / 'data' / 'raw' / '2023-2025'
omie_files = sorted(raw_dir.glob('*.txt'))

print("Validando datos OMIE (2023-2025)...")
print("="*50)

if not omie_files:
    print("[FAIL] No hay archivos OMIE en data/raw/2023-2025/")
else:
    first_date = datetime.strptime(omie_files[0].stem, '%Y%m%d').date()
    last_date = datetime.strptime(omie_files[-1].stem, '%Y%m%d').date()
    
    print(f"[OK] {len(omie_files):,} archivos encontrados")
    print(f"     Rango: {first_date} → {last_date}")
    
    if first_date.year == 2023 and last_date.year == 2025:
        print(f"[OK] Cobertura 2023-2025 confirmada")
```

## 4. Validación de datos ESIOS

```python
esios_raw_dir = project_root / 'data' / 'raw' / 'esios'

print("\nValidando estructura ESIOS...")
print("="*50)

if esios_raw_dir.exists():
    print(f"[OK] Directorio esios presente")
    print(f"     Los datos ESIOS se obtienen desde API REE (caché local)")
else:
    print("[INFO] Crear directorio si se necesita caché local")
```
